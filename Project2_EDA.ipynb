# DecodeLabs Data Analytics Internship

## Week 2: Exploratory Data Analysis (EDA)

### Objective
The objective of this project is to perform Exploratory Data Analysis (EDA) on a cleaned e-commerce dataset to understand data distributions, identify patterns, detect outliers, and generate business insights using Python.

**Tools Used**
- Python
- Pandas
- NumPy
- Matplotlib
- Seaborn
- Google Colab

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

plt.style.use('ggplot')
sns.set_style("whitegrid")
```

```python
df = pd.read_excel("dataset.xlsx")
df.head()
```

```python
print("Dataset Shape:", df.shape)
print("\nColumns:")
print(df.columns.tolist())
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
print("Duplicate Rows:", df.duplicated().sum())
```

```python
numeric_cols = ["Quantity", "UnitPrice", "ItemsInCart", "Revenue"]

for col in numeric_cols:
    print(f"\n----- {col} -----")
    print("Mean:", round(df[col].mean(), 2))
    print("Median:", round(df[col].median(), 2))
    print("Mode:", df[col].mode()[0])
```

```python
numeric_df = df.select_dtypes(include=['int64', 'float64'])

corr = numeric_df.corr()

corr
```

```python
plt.figure(figsize=(8,6))

sns.heatmap(corr,
            annot=True,
            cmap="coolwarm")

plt.title("Correlation Heatmap")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(df["Revenue"], bins=20, kde=True)

plt.title("Revenue Distribution")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(x=df["Revenue"])

plt.title("Revenue Boxplot")

plt.show()
```

```python
monthly = df.groupby("Month")["Revenue"].sum().sort_values(ascending=False)

plt.figure(figsize=(10,5))

monthly.plot(kind="bar")

plt.title("Monthly Revenue")

plt.xlabel("Month")

plt.ylabel("Revenue")

plt.xticks(rotation=45)

plt.show()
```

```python
product = df.groupby("Product")["Revenue"].sum().sort_values(ascending=False)

plt.figure(figsize=(12,5))

product.plot(kind="bar")

plt.title("Revenue by Product")

plt.xlabel("Product")

plt.ylabel("Revenue")

plt.xticks(rotation=45)

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(data=df, x="PaymentMethod")

plt.title("Payment Method Distribution")

plt.xticks(rotation=20)

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(data=df, x="OrderStatus")

plt.title("Order Status")

plt.xticks(rotation=20)

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.scatterplot(data=df,
                x="Quantity",
                y="Revenue")

plt.title("Quantity vs Revenue")

plt.show()
```
